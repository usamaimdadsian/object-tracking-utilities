```python
%load_ext autoreload
%autoreload 2
from utils.tracking_evaluator import TrackingEvaluator
import numpy as np
```

```python
data = np.load("data_track_1.npy")
data = data[:,:-1]
```

```python
evaluator = TrackingEvaluator(data)
evaluator.fit()
evaluator.predict()
print(evaluator.calc_error("mse"))
print(evaluator.calc_error("mae"))
print(evaluator.calc_error("custom"))
print(evaluator.accuracy())
evaluator.visualize()
```

```python
evaluator = TrackingEvaluator(data)
evaluator.fit()
evaluator.predict()

x_noise, y_noise = 5, 5
print(evaluator.calc_error("mse",x_noise=x_noise, y_noise=y_noise))
print(evaluator.accuracy(threshold=10,x_noise=x_noise, y_noise=y_noise))
evaluator.visualize(draw_noisy=True)
```
